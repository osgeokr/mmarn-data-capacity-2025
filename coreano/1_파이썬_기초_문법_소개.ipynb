# 산수 연산 기본

파이썬은 계산기처럼 사용할 수 있습니다.  
기본적인 연산은 다음과 같습니다.

- `+` 덧셈
- `-` 뺄셈
- `*` 곱셈
- `/` 나눗셈
- `**` 거듭제곱
- `%` 나머지

습관적으로 괄호를 넣어 계산 순서를 명확히 하면 좋습니다.

````python
a = 10
b = 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a ** b =", a ** b)
print("a % b =", a % b)
````

# 자료형과 변수

파이썬의 자료형은 데이터를 표현하는 가장 기본 단위입니다.  
초보자가 꼭 알아야 할 기본 자료형은 다음과 같습니다.

- 정수(int): 5, 20, -3  
- 실수(float): 3.14, 0.98, -12.5  
- 문자열(str): "Santo Domingo", "환경부", "온도"  
- 불(bool): True, False  

자료형은 환경 데이터를 다룰 때 매우 중요합니다.  
예를 들어 온도(°C)는 실수, 관측 지점 이름은 문자열, 기준 초과 여부는 불(bool)로 표현합니다.

변수(variable)는 데이터를 저장하는 이름표입니다.
파이썬에서는 `=` 기호로 변수를 만듭니다.

````python
# 관측값 예시
temperature = 28.5        # 온도(°C)
station_name = "Santo Domingo"  # 측정 지점명
rainfall = 12             # 강수량(mm)
is_protected_area = True  # 보호구역 여부

print(type(temperature), temperature)
print(type(station_name), station_name)
print(type(rainfall), rainfall)
print(type(is_protected_area), is_protected_area)
````

# 리스트(list)

리스트는 여러 개의 값을 한꺼번에 보관하는 자료형입니다.
환경 모니터링 데이터처럼 "여러 지점의 측정값"을 저장할 때 매우 유용합니다.

예시:
- 여러 지역의 온도 리스트
- 하루 동안의 시간별 강수량 리스트
- 여러 지점의 pH 값 목록

리스트 특징:
- 대괄호 [ ] 를 사용한다.
- 값에 순서가 있다 (0부터 시작).
- 길이는 len() 함수로 확인한다.

````python
# 여러 지역의 대기 온도(°C)
temperatures = [28.5, 29.1, 30.0, 27.8]

# 첫 번째 값 확인
print("첫번째 온도:", temperatures[0])

# 리스트에 값 추가
temperatures.append(29.4)

# 전체 길이
print("리스트 길이:", len(temperatures))

# 값 전체 출력
print("전체 목록:", temperatures)
````

# 딕셔너리(dict)

딕셔너리는 "이름: 값" 형태로 저장되는 자료형입니다.
환경 데이터에서 관측 지점 정보를 표현할 때 특히 유용합니다.

예:
- {"지점": "Santo Domingo", "온도": 28.5, "습도": 72}
- {"종": "황금제비", "개체수": 12, "보호종": True}

딕셔너리는 중괄호 { } 를 사용하고, 키(key)로 값을 가져옵니다.

````python
# 환경 관측 지점 정보
station = {
    "name": "Santo Domingo",
    "temperature": 28.5,
    "humidity": 72,
    "protected_area": True
}

print(station["name"])
print(station["temperature"])
print(station["protected_area"])
````

# 불(bool)과 비교 연산

불(bool)은 '참(True)'과 '거짓(False)' 두 가지 값을 가지는 자료형입니다.

환경 데이터에서는 이런 비교가 자주 등장합니다.
- 온도가 기준보다 높은가?
- 오염도가 허용치를 넘었는가?
- 강수량이 0보다 큰가?

비교 연산자:
- `==` 같다
- `!=` 다르다
- `>` 크다
- `<` 작다
- `>=` 크거나 같다
- `<=` 작거나 같다

````python
temperature = 32
threshold = 30

print("기준 초과 여부:", temperature > threshold)
print("같은 값인가?", temperature == threshold)
print("낮은가?", temperature < threshold)
````

# if 조건문

조건에 따라 프로그램이 다르게 움직이도록 만드는 문법입니다.

환경부 실무에서 자주 쓰이는 예:
- 온도가 30°C 이상이면 '고온 경보'
- 대기 오염 지수가 기준을 초과하면 '주의'
- 보호구역 여부에 따라 다른 처리 적용

기본 구조:
```
if 조건:
    실행문
elif 조건:
    실행문
else:
    실행문
```

````python
pm25 = 42  # 대기 중 PM2.5 농도(µg/m³)

if pm25 > 50:
    print("대기질: 나쁨 (주의 필요)")
elif pm25 > 25:
    print("대기질: 보통")
else:
    print("대기질: 좋음")
````

# for 반복문

리스트나 여러 데이터 값을 하나씩 꺼내서 반복 처리하는 문법입니다.

환경 데이터에서 매우 자주 사용됩니다.
예:
- 여러 지점의 온도를 하나씩 출력
- 1년 365일의 강수량 데이터를 순회
- 여러 종의 개체수를 하나씩 처리

````python
temperatures = [28.5, 29.1, 30.0, 27.8]

for t in temperatures:
    print("측정된 온도:", t)
````

# 함수 (Function)

여러 번 반복되는 코드를 하나로 묶어 이름을 붙인 것입니다.

환경 데이터에서 함수는 다음과 같은 작업을 자동화하는 데 유용합니다.
- 온도 기준 판단
- 대기 오염 등급 산출
- 값 단위 변환

````python
def check_temp(temp):
    if temp >= 30:
        return "고온"
    else:
        return "정상"

print(check_temp(28))
print(check_temp(32))
````

### pip란?
파이썬 패키지를 설치하는 도구입니다. pip을 사용하면 필요한 패키지를 PyPI에서 자동으로 다운로드해 설치합니다.

예:
pip install pandas
pip install geopandas

### PyPI란?
**PyPI (Python Package Index)** 는 전 세계 개발자들이 만든 파이썬 패키지를 모아둔 중앙 저장소입니다.

- PyPI 공식 주소: https://pypi.org/

pip은 기본적으로 PyPI에서 패키지를 다운로드하여 설치합니다.

즉,

- **pip = 파이썬 패키지 설치 도구**
- **PyPI = 파이썬 패키지가 보관되는 창고**

이 섹션에서는 파이썬 설치 → Jupyter 설치 → pip과 PyPI 사용 흐름을 소개했습니다.
